```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 生成具有线性可分特性的样本数据
np.random.seed(0)
num_samples = 50

# 生成两类样本数据，每类都有3个特征
class_0 = np.random.rand(num_samples, 3) * 5
class_1 = np.random.rand(num_samples, 3) * 5 + 6

# 创建一个绘图对象
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制两类样本的散点图
ax.scatter(class_0[:, 0], class_0[:, 1], class_0[:, 2], c='b', marker='o', label='Class 0')
ax.scatter(class_1[:, 0], class_1[:, 1], class_1[:, 2], c='r', marker='x', label='Class 1')

# 设置图形属性
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('Linearly Separable Samples with Separating Plane')
ax.legend()

# 绘制分隔平面，这里以一个简单的平面为例
normal_vector = np.array([1, -1, 1])  # 平面的法向量
point_on_plane = np.array([5, 5, 5])  # 平面上的一个点
xx, yy = np.meshgrid(np.arange(0, 10, 0.5), np.arange(0, 10, 0.5))
zz = (-normal_vector[0] * xx - normal_vector[1] * yy + point_on_plane[2]) / normal_vector[2]
ax.plot_surface(xx, yy, zz, color='gray', alpha=0.3)

# 保存图片
plt.savefig(
    '09th_img/img2.png',
    dpi=200,
    facecolor='white',
    edgecolor='red',
    bbox_inches='tight'
)
```

```python
# 导包
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 导入SVM对象
from sklearn.svm import SVC

# 导入数据集
from sklearn.datasets import load_iris

# 导入模型评估指标
from sklearn.metrics import accuracy_score, classification_report

# 导入数据集划分
from sklearn.model_selection import train_test_split

from IPython.core.interactiveshell import InteractiveShell # 这个对象设置所有行全部输出

# 设置该对象ast_node_interactivity属性的值为all，表示在notebook下每一行有输出的代码全部输出运算结果
InteractiveShell.ast_node_interactivity = "all" 

# 解决坐标轴刻度负号乱码
plt.rcParams['axes.unicode_minus'] = False

# 基本参数设置
plt.rcParams['font.sans-serif']=['Simhei']
plt.rcParams['axes.unicode_minus']=False

# 加载Iris数据集
iris = load_iris()
x = iris.data
y = iris.target

# 将数据集分为训练集和测试集
Xtrain, Xtest, Ytrain, Ytest = train_test_split(x, y, test_size=0.3, random_state=42)

# 创建SVC模型实例
clf = SVC(kernel='linear', C=1.0, random_state=42)

# 训练模型
clf.fit(Xtrain, Ytrain)

# 预测
y_pred = clf.predict(Xtest)

# 评估
accuracy_score(Ytest, y_pred)

# 输出分类报告
print("Classification Report:\n", classification_report(Ytest, y_pred))
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# 判断是gpu还是cpu
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 定义SVM模型   
class SVM(nn.Module):
    def __init__(self, input_size, num_classes):
        super(SVM, self).__init__()
        self.linear = nn.Linear(input_size, num_classes)

    def forward(self, x):
        return self.linear(x)

#定义训练数据
x_train = torch.tensor([[1., 1.], [-1., 1.], [-1., -1.], [1., -1.]]).to(device)
y_train = torch.tensor([1., -1., -1., -1.]).to(device)

# 定义SVM模型   
svm = SVM(input_size=2, num_classes=1).to(device)
criterion = nn.HingeEmbeddingLoss()
optimizer = optim.SGD(svm.parameters(), lr=0.01)

#训练模型
num_epochs = 1000
for epoch in range(num_epochs):
    optimizer.zero_grad()
    outputs = svm(x_train)
    loss = criterion(outputs.squeeze(), y_train)
    loss.backward()
    optimizer.step()
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")

#测试模型
x_test = torch.tensor([[2., 2.], [-2., 2.], [-2., -2.], [2., -2.]]).to(device)
outputs = svm(x_test)
predicted = torch.sign(outputs).squeeze().cpu()
print(predicted)
```
