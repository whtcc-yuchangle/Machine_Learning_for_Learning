```python
# 使用Spider抓取懂车帝数值型数据后打包为对应格式数据并使用pandas将数据写入到csv文件；
# Packages and Modules import statements
# --------------------------------------
import requests
import pandas as pd
import numpy as np
from lxml import etree


# codings
# --------------------------------------

class DongchediSpider(object):
    '''
    采集懂车帝车辆评分的数值型数据
    '''

    def __init__(self):
        self.__headers = {
            'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'
        }

    @property
    def headers(self):
        return self.__headers

    def __get_requests(self,url='',headers=None,method='GET',form_data=''):

        if method == 'GET':
            return requests.get(url=url,headers=headers)
        else:
            return requests.post(url=url,headers=headers,data=form_data)

    # 公共接口封装
    def get_requests(self,url='',headers=None,method='GET',form_data=''):

        return self.__get_requests(url=url,headers=headers,method=method,form_data=form_data)

    def __parse_data(self,data):
        '''
        解析一页数据
        :param data:str类型；
        :return: list类型
        '''

        dom = etree.HTML(data)
        comments = dom.xpath('//section[@class="tw-col-span-40 md:tw-col-span-29"]/article')
        page_data = []

        for comment in comments:
            # 字段数据依次解析

            # 作者id
            auther_id = comment.xpath(
                './aside[@class="tw-col-span-40 xl:tw-col-span-12 xl:tw-pr-12"]/header/div[@class="tw-ml-12 tw-flex tw-flex-col tw-justify-center"]/h2/a/text()')[
                0]

            # 总评分
            finall_score = comment.xpath(
                './section[@class="tw-col-span-40 tw-pt-16 xl:tw-col-span-28 tw-p-12 xl:tw-pt-0"]/header/a/span[1]/text()')[
                0]

            # 购车价格
            buy_car_price = '' if not comment.xpath('./aside/section/div[2]/div[1]/div[3]/p[1]/text()') else comment.xpath('./aside/section/div[2]/div[1]/div[3]/p[1]/text()')[0]

            # 续航公里数
            xuhang = '' if not comment.xpath('./aside/section/div[2]/div[1]/div[4]/p[1]/text()') else comment.xpath('./aside/section/div[2]/div[1]/div[4]/p[1]/text()')[0]

            # 一组核心的数值数据采集
            waiguan_score = '' if not comment.xpath('./aside/section/div[2]/div[2]/div[2]/p[1]/text()') else comment.xpath('./aside/section/div[2]/div[2]/div[2]/p[1]/text()')[0]

            neishi_score = '' if not comment.xpath('./aside/section/div[2]/div[2]/div[3]/p[1]/text()') else comment.xpath('./aside/section/div[2]/div[2]/div[3]/p[1]/text()')[0]

            peizhi_score = '' if not comment.xpath('./aside/section/div[2]/div[2]/div[4]/p[1]/text()') else comment.xpath('./aside/section/div[2]/div[2]/div[4]/p[1]/text()')[0]

            kongjian_score = '' if not comment.xpath('./aside/section/div[2]/div[2]/div[5]/p[1]/text()') else comment.xpath('./aside/section/div[2]/div[2]/div[5]/p[1]/text()')[0]

            shushixing_score = '' if not comment.xpath('./aside/section/div[2]/div[2]/div[6]/p[1]/text()') else comment.xpath('./aside/section/div[2]/div[2]/div[6]/p[1]/text()')[0]

            caokong_score = '' if not comment.xpath('./aside/section/div[2]/div[2]/div[7]/p[1]/text()') else comment.xpath('./aside/section/div[2]/div[2]/div[7]/p[1]/text()')[0]

            dongli_score = '' if not comment.xpath('./aside/section/div[2]/div[2]/div[8]/p[1]/text()') else comment.xpath('./aside/section/div[2]/div[2]/div[8]/p[1]/text()')[0]


            line = [auther_id,finall_score,waiguan_score,neishi_score,peizhi_score,kongjian_score,shushixing_score,caokong_score,dongli_score]
            page_data.append(line)

        return page_data

    def __cleaner_data(self,data):
        '''
        数据的简单清洗：将显示为'-'的数据置为0；将''的数据也置为0
        :param data:
        :return:
        '''

        for row in data:
            for col_index in range(len(row)):
                if row[col_index] == '-' or row[col_index] == '':
                    row[col_index] = 0

        return data

    def getDataFrame(self,data):
        '''
        将待写入数据构造为dataframe结构
        :return:
        '''

        data_frame = pd.DataFrame(data=data,columns=['作者id','总评分','外观评分','内饰评分','配置评分','空间评分','舒适性评分','操控评分','动力评分'])

        return data_frame

    def write_to_csv(self,dataframes):
        '''

        :param dataframes:
        :return:
        '''

        dataframes.to_csv('./02th_dongchediData.csv',sep=';')
        
    def display(self):

        url = 'https://www.dongchedi.com/auto/series/score/4363-x-S0-x-x-x-{}'

        pages = int(input('请输入采集数据的页数>>>：'))

        final_result = []
        print('开始采集数据...')
        for i in range(1,pages + 1):

            url = url.format(i)

            # 请求一页数据
            html = self.__get_requests(url=url,headers=self.__headers).text

            # 解析一页数据
            page_data = self.__parse_data(html)

            # 清洗数据
            final_page_data = self.__cleaner_data(page_data)
            for row in final_page_data:
                final_result.append(row)
            # print(page_data)

        print('开始数据写入...')
        # 最后的数据写入
        self.write_to_csv(self.getDataFrame(final_result))

def test():
    DongchediSpider().display()



# run test UseCase if current modules is main
if __name__ == '__main__':
    test()
```
